```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
x = np.linspace(0, 4, 12)
y = np.cos(x**2/3+4)
#plt.plot(x,y,'o')
#plt.show()
coeff1 = np.polyfit(x,y,1)
coeff2 = np.polyfit(x,y,2)
coeff3 = np.polyfit(x,y,3)
coeff4 = np.polyfit(x,y,4)
print(coeff1)
print(coeff2)
print(coeff3)
print(coeff4)
plt.plot(x,y,'o')
plt.plot(x,np.polyval(coeff1,x),'black')
plt.plot(x,np.polyval(coeff2,x),'green')
plt.plot(x,np.polyval(coeff3,x),'red')
plt.plot(x,np.polyval(coeff4,x),'cyan')
plt.legend(['data', '1st degree poly', '2nd degree poly','3rd degree poly','4th degree poly'],loc = 'best')
plt.show()
```
